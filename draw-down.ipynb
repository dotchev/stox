```python
import pandas as pd
import yfinance as yf
from datetime import date
```

```python
# Define the ticker symbol
tickerSymbol = 'IYW'

# Get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

# Get the historical prices for this ticker
history = tickerData.history(interval='1wk', start='2021-1-1', end=date.today())
```

```python
weeks = 4
history['WLow'] = history.Low.rolling(window=pd.api.indexers.FixedForwardWindowIndexer(window_size=weeks)).min().shift(-1)
history['DrawDown'] = history.WLow / history.Close - 1
print(f'{tickerSymbol} {weeks}w median draw down {history.DrawDown.median():.1%}')
display(history)
```
