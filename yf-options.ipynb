```python
import yfinance as yf
import pandas as pd

symbol = "NVDA"  # Example stock symbol, you can change it to any valid ticker
print(symbol)

# Create a ticker object
ticker = yf.Ticker(symbol)  # Example with Apple
info = ticker.info
display(info)
```

```python

# Get available expiration dates
expiration_dates = ticker.options
print(f"Available expiration dates for {symbol}:")
display(expiration_dates)
```

```python

expiration = expiration_dates[0]
options_chain = ticker.option_chain()
display(options_chain)
```

```python

# options_chain contains two DataFrames: calls and puts
print("expiration:", expiration)
display(options_chain.calls)
```
