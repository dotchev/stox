```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Define the ticker symbol
tickerSymbol = 'SMH'

# Get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

# Get the historical prices for this ticker
tickerDf = tickerData.history(interval='1d', start='2021-1-1', end='2024-5-5')

```

```python
# See your data
tickerDf.head()
```

```python
# Calculate the 20 and 50 days moving averages of the closing prices
tickerDf['MA200'] = tickerDf['Close'].rolling(window=200).mean()
tickerDf['MA50'] = tickerDf['Close'].rolling(window=50).mean()
```

```python
# Plot everything by leveraging the very powerful matplotlib package
plt.figure(figsize=(16,9))
plt.plot(tickerDf['Close'], label=tickerSymbol)
plt.plot(tickerDf['MA200'], label='MA 200 days')
plt.plot(tickerDf['MA50'], label='MA 50 days')
# plt.axhline(y=0, color='k', linestyle='--')
plt.title(f'{tickerSymbol} Stock Price History')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend(loc='upper left')
plt.show()
```
