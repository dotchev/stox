```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from statistics import mean
from datetime import date
from yfetch import get_stock_history
from symbols import etfs_with_weekly_options, mags_weeks, my_picks
```

```python
weeks = 1
rows = []
symbols = my_picks
for symbol in symbols:
    history = get_stock_history(symbol, interval='1wk').tail(13)
    changes = history.Close.pct_change(periods=weeks)
    change = changes.clip(lower=0).mean()
    std = changes.std()
    rows.append({
        'symbol': symbol,
        'weeks': len(history),
        'change': change,
        'median': changes.median(),
        'std': std,
        'ratio': change / std if std != 0 else 0,
    })

df = pd.DataFrame(rows).sort_values(by='ratio', ascending=False).reset_index(drop=True)
df.index += 1
df.to_csv('data/stocks-1w.csv', index=False)
for col in ['change', 'median', 'std', 'ratio']:
    df[col] = df[col].map('{:.2%}'.format)
display(df)
```
