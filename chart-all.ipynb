```python
import os
import matplotlib.pyplot as plt
from file import load_file
```

```python
plt.figure(figsize=(16,6))

for symbol in os.listdir('data'):
    d = load_file('data/'+symbol)
    option_weeks = [p[0] for p in d]
    option_prof = [p[1] for p in d]
    plt.plot(option_weeks, option_prof, label=symbol, marker='.')

plt.legend(loc='upper right')
plt.title('Option profit')
plt.xlabel('weeks')
plt.ylabel('%')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,6))

for symbol in os.listdir('data'):
    d = load_file('data/'+symbol)
    option_weeks = [p[0] for p in d]
    option_positive = [p[2] for p in d]
    plt.plot(option_weeks, option_positive, label=symbol, marker='.')

plt.legend(loc='upper left')
plt.title('Option positive')
plt.xlabel('weeks')
plt.ylabel('%')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,6))

for symbol in os.listdir('data'):
    d = load_file('data/'+symbol)
    option_weeks = [p[0] for p in d]
    option_prof = [p[1] for p in d]
    option_positive = [p[2] for p in d]
    option_score = [x * y / 100 for x, y in zip(option_prof, option_positive)]
    plt.plot(option_weeks[:10], option_score[:10], label=symbol, marker='.')

plt.legend(loc='upper right')
plt.title('Option score (gain * positive)')
plt.xlabel('weeks')
plt.ylabel('%')
plt.grid()
plt.show()
```
